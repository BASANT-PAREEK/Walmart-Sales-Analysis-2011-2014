<H4> OVERALL SALES TREND

```python
#Getting month year from the dataset
data['month_year']=data['Order Date'].apply(lambda x: x.strftime('%Y-%m'))
data['month_year']
```

```python
#Grouping month year
data_trend=data.groupby('month_year').sum()['Sales'].reset_index()
```

```python
plt.figure(figsize=(15,6))
plt.plot(data_trend['month_year'],data_trend['Sales'],color='black')
plt.xticks(rotation='vertical',size=8)
plt.tight_layout()
plt.show()
```

<H4> MOST SELLING PRODUCT(BY SALES)

```python
#GROUPING PRODUCT NAME Column
prod_sales = pd.DataFrame(data.groupby('Product Name').sum()['Sales'])
```

```python
#SORTING prod_sales column
prod_sales = prod_sales.sort_values('Sales',ascending=False)
```

```python
#TOP 10 PRODUCTS BY SALES
prod_sales[:10]
```

<h4> MOST SELLING PRODUCTS (BY QUANTITY)
